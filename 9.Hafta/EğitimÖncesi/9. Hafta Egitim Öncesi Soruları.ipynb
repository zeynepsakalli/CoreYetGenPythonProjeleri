"init" bir Python sınıfının yapıcı (constructor) yöntemidir. Yeni bir nesne oluşturulduğunda otomatik olarak çağrılır ve nesnenin başlatılmasını sağlar.
Yani, bir sınıfın init yöntemi, bir nesnenin oluşturulduğu zamanda yapılması gereken tüm işlemleri gerçekleştirir. Örneğin, bir sınıfın özelliklerinin (attribute) başlangıç değerlerinin atanması, diğer nesnelerle ilişkilerin kurulması veya bazı önemli verilerin yüklenmesi gibi işlemler yapılabilir.
Özetle, Python'da init yöntemi, bir sınıfın nesnesinin oluşturulması sırasında çalışan özel bir yöntemdir ve sınıfın nesnesinin oluşturulmasından önce yapılması gereken tüm hazırlıkları gerçekleştirir.

"self", bir Python sınıfında bir yöntemin ilk argümanıdır ve o yöntemi çağıran nesneyi temsil eder.
Yani, bir sınıfın bir yöntemi çağrıldığında, Python otomatik olarak o yönteme ait nesnenin referansını "self" adı verilen ilk parametreye atar. Bu sayede, o yöntem içerisinde o nesnenin özelliklerine ve yöntemlerine kolayca erişilebilir.
Örneğin, bir sınıfın "isim" özelliğine "self.isim" şeklinde erişilebilir ve bir yöntem içerisinde "self.yöntem()" şeklinde başka bir yöntem çağrılabilir.
Ayrıca, sınıfın farklı nesneleri farklı değerlere sahip olabileceğinden, "self" parametresi, farklı nesnelerin özelliklerine ve yöntemlerine erişebilmek için kullanılır.
Özetle, "self", bir sınıfın yöntemlerinde o yöntemi çağıran nesneye erişmek için kullanılan özel bir parametredir.


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Bus:
    def __init__(self, model, speed, year):
        self.speed = speed
        self.model = model
        self.year = year

    def seating_capacity(self, capacity=50):
        return f"The seating capacity of a bus is {capacity} passengers."

bus = Bus("Mercedes", "Sprinter", 2021)
print(bus.seating_capacity()) # The seating capacity of a bus is 50 passengers.

````
