````python
class Person:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
        
    def intro(self):
        print(self.name, self.surname, self.age)

class Student(Person):
    def __init__(self, name, surname, age, number):
        super().__init__(name, surname, age)
        self.number = number
        
    def intro(self):
        print(self.name, self.surname, self.age, self.number)

    def study(self):
        print(f"{self.number} numaralı öğrenci ders çalışıyor.")

class Teacher(Person):
    def __init__(self, name, surname, age, branch):
        super().__init__(name, surname, age)
        self.branch = branch
        
    def intro(self):
        print(self.name, self.surname, self.age, self.branch)

    def teach(self):
        print(f"{self.name} isimli öğretmenin {self.branch} dersine giriyor.")

p1 = Person("Zeynep", "Sakallı", 21)
p1.intro()

s1 = Student("Ali", "Yılmaz", 18, 12345)
s1.intro()
s1.study()

t1 = Teacher("Ayşe", "Demir", 35, "Matematik")
t1.intro()
t1.teach()
````

````python
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def speed(self, speed): ##set_speed kullanırsak daha güvenli olabilirmiş.
        self.speed = speed
        
    def mileage(self, mileage): ##set_mileage kullanılması daha güvenli
        self.mileage = mileage
        
    def __str__(self):
        return f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}"       

class Bus(Vehicle):
    pass

bus = Bus("School Volvo", 180, 12)
print(bus)  # Name: School Volvo Speed: 180 Mileage: 12
````
