# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buaya Yazınız
def vki_hesapla(kilo, boy):
    vki = kilo / (boy ** 2)
    return vki

def vki_siniflandirma(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez"

kullanici_sayisi = int(input("Kaç kullanıcı girişi yapacaksınız? "))

kullanicilar = []

for i in range(kullanici_sayisi):
    print(f"\nKullanıcı {i+1}")
    ad_soyad = input("Adınız ve soyadınız: ")
    yas = int(input("Yaşınız: "))
    boy = float(input("Boy (metre cinsinden): "))
    kilo = float(input("Kilo (kg cinsinden): "))

    vki = vki_hesapla(kilo, boy)
    sinif = vki_siniflandirma(vki)

    kullanici = {
        "Ad Soyad": ad_soyad,
        "Yaş": yas,
        "Boy": boy,
        "Kilo": kilo,
        "VKI": vki,
        "Sınıf": sinif
    }

    kullanicilar.append(kullanici)

# Verileri dosyaya kaydetme
dosya_adi = "kullanici_verileri.txt"

with open(dosya_adi, "w") as dosya:
    for kullanici in kullanicilar:
        dosya.write("Ad Soyad: " + kullanici["Ad Soyad"] + "\n")
        dosya.write("Yaş: " + str(kullanici["Yaş"]) + "\n")
        dosya.write("Boy: " + str(kullanici["Boy"]) + "\n")
        dosya.write("Kilo: " + str(kullanici["Kilo"]) + "\n")
        dosya.write("VKI: " + str(kullanici["VKI"]) + "\n")
        dosya.write("Sınıf: " + kullanici["Sınıf"] + "\n")
        dosya.write("\n")

print("Veriler dosyaya kaydedildi.")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime

class SmoothieBank:
    def __init__(self):
        self.menu = {
            "1": {"name": "Mango Smoothie", "price": 10},
            "2": {"name": "Çilek Smoothie", "price": 12},
            "3": {"name": "Ananas Smoothie", "price": 15},
            "4": {"name": "Muz Smoothie", "price": 8},
            "5": {"name": "Elma Smoothie", "price": 9},
        }
        self.transactions = []

    def show_menu(self):
        print("=== Menü ===")
        for key, value in self.menu.items():
            print(f"{key}. {value['name']}: {value['price']} TL")
        print()

    def purchase(self):
        self.show_menu()
        choice = input("Satın almak istediğiniz smoothie'nin numarasını girin (Çıkış için 'q' tuşuna basın): ")
        
        while choice != 'q':
            if choice in self.menu:
                smoothie = self.menu[choice]
                self.transactions.append((datetime.datetime.now(), smoothie['name'], smoothie['price']))
                print(f"{smoothie['name']} satın alındı. Fiyat: {smoothie['price']} TL")
            else:
                print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin.")
            choice = input("Satın almak istediğiniz smoothie'nin numarasını girin (Çıkış için 'q' tuşuna basın): ")

    def show_transactions(self):
        if len(self.transactions) > 0:
            print("=== Satın Alma Geçmişi ===")
            for transaction in self.transactions:
                date = transaction[0]
                smoothie_name = transaction[1]
                price = transaction[2]
                print(f"Tarih: {date} | Smoothie: {smoothie_name} | Fiyat: {price} TL")
        else:
            print("Henüz bir satın alma işlemi yapılmamış.")

smoothie_bank = SmoothieBank()

while True:
    print("=== YetBar - Smoothie Bank ===")
    print("1. Menüyü Görüntüle")
    print("2. Satın Al")
    print("3. Satın Alma Geçmişini Görüntüle")
    print("4. Çıkış")
    choice = input("Bir işlem seçin: ")

    if choice == "1":
        smoothie_bank.show_menu()
    elif choice == "2":
        smoothie_bank.purchase()
    elif choice == "3":
        smoothie_bank.show_transactions()
    elif choice == "4":
        break
    else:
        print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin.")

print("Programdan çıkılıyor...")
```
